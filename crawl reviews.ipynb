```python
from bs4 import BeautifulSoup

import urllib.request
import csv
import json
import pandas as pd
```

```python
urlpage = 'https://www.imdb.com/title/tt9898836/reviews'

page = urllib.request.urlopen(urlpage)
soup = BeautifulSoup(page, 'html.parser')
```

```python
review_title_bs = soup.find_all('a', attrs={'class': 'title'})

review_title = list()

for r1 in review_title_bs:
    review_title.append(r1.getText().replace('\n', ''))
```

```python
reviews_bs = soup.find_all('div', attrs={'class': 'text show-more__control'})

reviews = list()

for r in reviews_bs:
    reviews.append(r.getText())
```

```python
review_dates_bs = soup.find_all('span', attrs={'class': 'review-date'})

review_dates = list()

for rd in review_dates_bs:
    review_dates.append(rd.getText())
```

```python
ind_rating_bs = soup.find_all('span', attrs={'class': 'rating-other-user-rating'})

ind_rating = list()

for ind in ind_rating_bs:
    ind_rating.append(int(ind.span.getText()))
```

```python
spoiler_bs = soup.find_all('div', attrs={'class': 'lister-item-content'})

spoiler = list()

for s in spoiler_bs:
    if s.getText().find('Warning: Spoilers') > 1:
        # print('find spoiler')
        spoiler.append('yes')
    else:
        # print('not spoiler')
        spoiler.append('no')
```

```python
df_review = pd.DataFrame(columns = ['review date', 'review title', 'review', 'rating', 'spoiler'])

df_review['review date'] = review_dates
df_review['review title'] = review_title
df_review['review'] = reviews
df_review['rating'] = ind_rating
df_review['spoiler'] = spoiler

df_review
```
