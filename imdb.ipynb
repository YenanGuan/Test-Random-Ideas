```python
from bs4 import BeautifulSoup

import urllib.request
import csv
import json
```

```python
urlpage = 'https://www.imdb.com/title/tt2560140/episodes?year=2019'
```

```python
page = urllib.request.urlopen(urlpage)
soup = BeautifulSoup(page, 'html.parser')
```

```python
print(soup)
```

```python
test = soup.find_all('div', attrs={'class': 'item_description'})
# results = test.find_all('body')
```

```python
for t in test:
    reviews = t.getText()
    print(reviews)
```

```python
test = soup.find_all('span', attrs={'class': 'ipl-rating-star__rating'})
# results = test.find_all('script')
```

```python
episodes_bs = soup.find_all('a', attrs={'itemprop': 'name'})

episodes = list()
for e in episodes_bs:
    episodes.append(e.get('title'))
    
episodes
```

```python
votes_bs = soup.find_all('span', attrs={'class': 'ipl-rating-star__total-votes'})

votes = list()

for v in votes_bs:
    temp = v.getText()
    temp = temp.replace('(', '')
    temp = temp.replace(')', '')
    temp = temp.replace(',', '')
    votes.append(int(temp))

votes
```

```python
rating_bs = soup.find_all('span', attrs={'class': 'ipl-rating-star__rating'})

rating = list()
for r in rating_bs:
    temp = r.getText()
    try:
        float(temp)
    except ValueError:
        pass
    else:
        if len(temp) == 3:
            rating.append(float(temp))
rating
```
